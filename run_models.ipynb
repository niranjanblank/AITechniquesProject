```python
from utilities import visualize_reconstruction, get_transform
from models import Autoencoder
import torch
```

```python
# path of the model
model_path = "trained_models/autoencoder_model_2.pth"
```

```python
# setting the device based on availability
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
model = Autoencoder().to(device)
```

```python
model.load_state_dict(torch.load(model_path))
```

```python
# get the transform to apply to the images
transform = get_transform()
```

```python
# testing normal image
# normal image
normal_image = "data/SmallerDataset/Test/Normal_Videos_912_x264_340.png"
reconstruction_error = visualize_reconstruction(model,  normal_image, transform, device=device)
print(f"Reconstruction Error: {reconstruction_error}")
```

```python
# testing anomaly
# anomaly images
anomaly_1 = "data/SmallerDataset/AnomalyImages/Arson010_x264_1460.png"
reconstruction_error = visualize_reconstruction(model,  anomaly_1, transform, device=device)
print(f"Reconstruction Error: {reconstruction_error}")
```
